# Ordinary least square Method

```python
%matplotlib inline
import matplotlib.pyplot as plt
Psize=[[6],[8],[10],[14],[18]]
Pprice=[7,9,13,17.5,18]
plt.figure()
plt.title('Pizza price plotted against diameter')
plt.xlabel('Diameter in inches')
plt.ylabel('Price in colors')
plt.plot(Psize,Pprice,'go')
plt.axis([0,25,0,25])
plt.grid(True)
```

```python
from sklearn.model_selection import train_test_split
PSize_train,PSize_test,PPrice_train,PPrice_test=train_test_split(Psize,Pprice,test_size=0.2)
plt.figure()
plt.title('Pizza price plotted against diameter')
plt.xlabel('Diameter in inches')
plt.ylabel('Price in dollars')
plt.plot(PSize_train,PPrice_train,'go')
plt.plot(Psize,Pprice,'x')
plt.axis([0,25,0,25])
plt.grid(True)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(PSize_train,PPrice_train)
print(PPrice_test)
model.predict(PSize_test)
```

```python
plt.figure()
plt.title('Pizza price plotted against diameter')
plt.xlabel('Diameter in inches')
plt.ylabel('Price in dollars')
plt.plot(PSize_train,PPrice_train,'go')
plt.plot(PSize_test,PPrice_test,'bo')
plt.axis([0,25,0,25])
plt.grid(True)

regression_line_y=model.predict([[0],[25]])
print(regression_line_y)
plt.plot([0,25],regression_line_y,color='r',linestyle='-',linewidth=1)
```

```python
model.predict(15)
```

```python
acc=model.score(Psize,Pprice)
print("Model Accuracy:",acc)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(Pprice,model.predict(Psize))
print('SGD MSE:',mse)
```
